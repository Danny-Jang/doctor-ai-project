```
from keras import optimizers
from keras_preprocessing.image import ImageDataGenerator

test_generator = ImageDataGenerator(rescale = 1./255)
test_set = test_generator.flow_from_directory('/content/drive/My Drive/skin_cancer/increased_data2/test',
                                            target_size = (224, 224),
                                            batch_size = 64,
                                            class_mode = 'categorical',
                                            shuffle=False)
```

```
from keras.models import load_model
model = load_model('/content/drive/My Drive/skin_cancer/skin_cancer_best_model3.h5')
```

```
scores = model.evaluate(test_set, steps = test_set.samples/test_set.batch_size)
print("%s: %.2f%%" %(model.metrics_names[1], scores[1]*100))
```

```
import numpy as np
import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)  # 모든 넘파이 어레이 보기 


output = model.predict_generator(
            test_set, 
            steps = test_set.samples/test_set.batch_size)
np.set_printoptions(formatter={'float': lambda x: "{0:0.3f}".format(x)})

print(output)
```

```
import numpy as np

test_set.reset()
predictions = model.predict(test_set, steps=test_set.samples/test_set.batch_size,verbose=1)
y_pred= np.argmax(predictions, axis=1)

print(y_pred)
```

```
y_test=np.array([])
for i in range(663):
    y_test=np.append(y_test,0)
for i in range(602):
    y_test=np.append(y_test,1)
print(y_test)
```

```
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, ConfusionMatrixDisplay
# labels = ['benign', 'malignant']
cm= confusion_matrix(y_test,y_pred)
print(cm)
```

```
ConfusionMatrixDisplay(cm, display_labels=('benign', 'malignant')).plot(values_format = 'd' ,cmap = 'GnBu')
```
